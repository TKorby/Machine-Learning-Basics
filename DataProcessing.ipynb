# Clean and Setup

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Ted\Documents\Mod9\data_main.csv')
```

```python
orig_shape = df.shape
orig_shape
```

```python
# change year and month to type int
months = {"Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun":6, "Jul":7, "Aug":8, "Sep":9, "Oct":10, "Nov":11, "Dec":12}
df['Month'].replace(months, inplace=True)
df["Month"] = df["Month"].astype(int)
```

```python
# df.drop('Index', axis=1, inplace=True)
df.info()
```

```python
# remove rows and columns with more than half values filled with nulls/NaN
df.dropna(thresh=orig_shape[1]/2, inplace=True)
print(df.info())
```

```python
df.dropna(thresh=orig_shape[0]/2, axis=1, inplace=True)
print(df.info())
```

```python
df.drop('Index', axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# change Year to int32 after fixing bad values: 92 -> 1992, nan -> removed
df["Year"].replace(92,1992, inplace=True)
df["Year"].unique()
```

```python
# remove commas from Length Duration and change to type int32
df["Length (Duration)"].replace(',','', regex=True, inplace=True)
df["Length (Duration)"] = pd.to_numeric(df["Length (Duration)"])
df["Length (Duration)"].describe()
print(df.info())
```

```python
df['Month'] = pd.to_numeric(df['Month'])
```

```python
print(df.info())
```

```python
df.describe()
```

```python
df["Year"].unique()
```

```python
for col in df.describe().columns:
    col_mean = df[col].mean()
    col_std = df[col].std()
    
    print(col, 'outliers')
    
    for value in df[col]:
        if abs(value - col_mean) > 3 * col_std:
            print(f'{value}, Z-score: {(value - col_mean) / col_std}')
    print('\r')
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
for col in df:
    print(f"{col} mising values: {df[col].isnull().sum()}")
```

```python
continuous_columns = ['Beats Per Minute (BPM)','Energy','Loudness (dB)','Liveness','Length (Duration)','Acousticness','Speechiness','Popularity']
```

```python
continuous_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
continuous_imputed_df = pd.DataFrame(continuous_imputer.fit_transform(df[continuous_columns]), columns= continuous_columns)
continuous_imputed_df
```

```python
df.info()
```

```python
for col in continuous_columns:
    df[col] = continuous_imputed_df[col]
df.info()
```

```python
categorical_columns = ['Artist','Top Genre','Year']
```

```python
categorical_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
categorical_imputed_df = pd.DataFrame(categorical_imputer.fit_transform(df[categorical_columns]), columns=categorical_columns)
categorical_imputed_df
```

```python
for col in categorical_columns:
    df[col] = categorical_imputed_df[col]
```

```python
for col in df:
    print(f"{col} missing values: {df[col].isnull().sum()}")
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df['date'] = df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-' + '1'
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['age'] = pd.to_datetime('2022-01-18') - df['date']
```

```python
age = pd.Series([df['age'][i].days for i in range(len(df['age']))])
```

```python
df['age'] = age
```

```python
df.drop(['Year', 'Month', 'date'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Top Genre'], drop_first=True)
```
